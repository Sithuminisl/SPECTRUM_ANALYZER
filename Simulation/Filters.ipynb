```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.random.randint(0, 10, 1000)
arr0 = np.arange(0, 1000, 1)
arr1 = (np.sin(arr0/10) + np.cos(arr0/20) + np.random.normal(0, 0.1, 1000))*10
arr3 = (np.sin(arr0/10) + np.cos(arr0/20))*10
arr2 = np.zeros(1000)

arr2[0] = arr1[0]

for i in range(1,1000):
    arr2[i] = (arr2[i-1] + arr1[i])//2

plt.plot(arr1[100:120])
plt.plot(arr2[100:120])
plt.show()
```

```python
f1 = plt.figure(figsize=(25, 10))
f1.add_subplot(111)
plt.plot(arr1[500:750])
plt.plot(arr2[500:750])
plt.plot(arr3[500:750])
```

```python
n = 7
N = 2**n

arr_fft = np.fft.fft(arr1[:N])
arr_fft = np.abs(arr_fft)/N
arr_fft = np.fft.fftshift(arr_fft)
f_values = np.arange(-N//2, N//2, 1)
plt.stem(f_values,arr_fft)
plt.grid()
```
